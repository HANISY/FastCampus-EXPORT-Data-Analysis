 (4) Outlier ÌôïÏù∏ (‚Äª Ï†ïÏÉÅÏ†ÅÏù∏ Î≤îÏ£ºÎ•º Î≤óÏñ¥ÎÇú Data)

````python
#Í∞Å Ïª¨ÎüºÎ≥Ñ ÏöîÏïΩ ÌÜµÍ≥ÑÎüâ ÌôïÏù∏
df.describe()
````

````python
# Ïª¨Îüº Î≤îÏúÑÎ•º ÏÑ†ÌÉùÌïòÏó¨ Ïù¥ÏÉÅÏπò Ïó¨Î∂Ä ÌôïÏù∏
selected_columns = df.columns[1:591]  # 0~589 ÏÑºÏÑúÎç∞Ïù¥ÌÑ∞ ÏÑ†ÌÉù

# IQRÏùÑ Í≥ÑÏÇ∞ÌïòÎäî Ìï®Ïàò Ï†ïÏùò
def detect_outliers_iqr(series, threshold=1.5):
    """
    IQR (Interquartile Range)ÏùÑ Ïù¥Ïö©ÌïòÏó¨ Ïù¥ÏÉÅÏπòÎ•º ÌÉêÏßÄÌïòÎäî Ìï®Ïàò
    :param series: Îç∞Ïù¥ÌÑ∞ ÏãúÎ¶¨Ï¶à
    :param threshold: Ïù¥ÏÉÅÏπòÎ°ú ÌåêÎã®ÌïòÎäî Í∏∞Ï§ÄÍ∞í (Í∏∞Î≥∏Í∞í: 1.5)
    :return: Ïù¥ÏÉÅÏπòÏùò Ïù∏Îç±Ïä§ Î¶¨Ïä§Ìä∏
    """
    # 1ÏÇ¨Î∂ÑÏúÑÏàòÏôÄ 3ÏÇ¨Î∂ÑÏúÑÏàò Í≥ÑÏÇ∞
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    # IQR Í≥ÑÏÇ∞
    IQR = Q3 - Q1
    # Ïù¥ÏÉÅÏπò Í≤ΩÍ≥Ñ Í≥ÑÏÇ∞
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    # Ïù¥ÏÉÅÏπò ÌÉêÏßÄ
    outliers = (series < lower_bound) | (series > upper_bound)
    return outliers

# ÏÑ†ÌÉùÎêú Ïª¨Îüº Î≤îÏúÑÏùò Ïù¥ÏÉÅÏπò Ïó¨Î∂Ä ÌôïÏù∏
outliers = df[selected_columns].apply(detect_outliers_iqr)

# Í∞Å Ïª¨ÎüºÏùò Ïù¥ÏÉÅÏπò ÎπÑÏú® Í≥ÑÏÇ∞
outlier_ratios = outliers.mean() * 100  # Ïù¥ÏÉÅÏπò ÎπÑÏú®Ïóê 100ÏùÑ Í≥±ÌïòÏó¨ ÌçºÏÑºÌä∏Î°ú Î≥ÄÌôò

# Ïù¥ÏÉÅÏπò ÎπÑÏú® Ï∂úÎ†•
print("Í∞Å ÏÑºÏÑúÎç∞Ïù¥ÌÑ∞ Ïª¨ÎüºÏùò Ïù¥ÏÉÅÏπò ÎπÑÏú® (%):")
print(outlier_ratios)
````

````python
# Ïù¥ÏÉÅÏπò ÎπÑÏú®ÏùÑ ÎÇ¥Î¶ºÏ∞®ÏàúÏúºÎ°ú Ï†ïÎ†¨ÌïòÏó¨ Ï∂úÎ†•
outlier_ratios_sorted = outlier_ratios.sort_values(ascending=False)

# Ïù¥ÏÉÅÏπò ÎπÑÏú® Ï∂úÎ†• (ÎÇ¥Î¶ºÏ∞®Ïàú)
print("Í∞Å ÏÑºÏÑúÎç∞Ïù¥ÌÑ∞ Ïª¨ÎüºÏùò Ïù¥ÏÉÅÏπò ÎπÑÏú® (%), ÎÇ¥Î¶ºÏ∞®Ïàú:")
print(outlier_ratios_sorted)
````

````python
from imblearn.over_sampling import SMOTE

# Pass/FailÏùò ÎπÑÏú® ÌôïÏù∏
class_distribution = df2['Pass/Fail'].value_counts()
print("Original Class Distribution:")
print(class_distribution)

# ÏûÖÎ†• Î≥ÄÏàòÏôÄ ÌÉÄÍ≤ü Î≥ÄÏàò Î∂ÑÎ¶¨
X = df2.drop(['Pass/Fail'], axis=1)       # Î∂àÎüâÏó¨Î∂ÄÏôÄ ÏÉÅÍ¥ÄÏÑ±Ïù¥ ÎÜíÏïòÎçò 14Í∞úÏùò ÏÑºÏÑú Îç∞Ïù¥ÌÑ∞ ÎåÄÏÉÅÏúºÎ°ú Î™®Îç∏ÎßÅ ÏßÑÌñâ
y = df2['Pass/Fail']

# ÌïôÏäµ Îç∞Ïù¥ÌÑ∞ Î∞è ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Î°ú Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2024)

# SMOTEÎ•º ÏÇ¨Ïö©ÌïòÏó¨ ÌÅ¥ÎûòÏä§ Î∂àÍ∑†Ìòï Î≥¥Ï†ï
smote = SMOTE(sampling_strategy='auto', random_state=2024)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Î≥¥Ï†ïÎêú ÌÅ¥ÎûòÏä§ Î∂ÑÌè¨ ÌôïÏù∏
resampled_class_distribution = pd.Series(y_train_resampled).value_counts()
print("\nResampled Class Distribution:")
print(resampled_class_distribution)
````

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Î°úÏßÄÏä§Ìã± ÌöåÍ∑Ä Î™®Îç∏ ÏÉùÏÑ±
model_lr = LogisticRegression(random_state=10)

# Î°úÏßÄÏä§Ìã± ÌöåÍ∑Ä Î™®Îç∏ ÌïôÏäµ
model_lr.fit(X_train_resampled, y_train_resampled)

# Î™®Îç∏ ÌèâÍ∞Ä
y_pred_lr = model_lr.predict(X_test)

# Ï†ïÌôïÎèÑ Î∞è Î∂ÑÎ•ò Î≥¥Í≥†ÏÑú Ï∂úÎ†•
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print(f"Accuracy of Logistic Regression Model: {accuracy_lr}")

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_lr))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_lr))
````

````python
from sklearn.inspection import permutation_importance

# Permutation Importance Í≥ÑÏÇ∞
# ÌäπÏÑ±Í∞íÏùÑ ÏàúÏó¥Î°ú ÎßåÎì† ÌõÑ Î™®Îç∏Ïùò ÏòàÏ∏° Ïò§Ï∞® Ï¶ùÍ∞ÄÎüâÏùÑ Í≥ÑÏÇ∞ÌïòÏó¨ ÌäπÏÑ±Í∞íÏùò Ï§ëÏöîÎèÑÎ•º Ï∏°Ï†ï
result = permutation_importance(model_lgbm, X_test, y_test, n_repeats=10, random_state=42, n_jobs=-1)

# ÌäπÏÑ± Ï§ëÏöîÎèÑÎ•º DataFrameÏúºÎ°ú Î≥ÄÌôòÌïòÏó¨ Ï∂úÎ†•
importance_df_permutation = pd.DataFrame({'Feature': X.columns, 'Importance': result.importances_mean})
importance_df_permutation = importance_df_permutation.sort_values(by='Importance', ascending=False)
print(importance_df_permutation)

# ÏÉÅÏúÑ 10Í∞ú ÌäπÏÑ± Ï§ëÏöîÎèÑ Ï∂îÏ∂ú
top_features_permutation = importance_df_permutation.head(10)

# Í∑∏ÎûòÌîÑ ÏÉùÏÑ±
plt.figure(figsize=(10, 6))
plt.barh(top_features_permutation['Feature'], top_features_permutation['Importance'], color='lightgrey')
plt.xlabel('Importance')
plt.title('Top 10 Permutation Feature Importances')
plt.gca().invert_yaxis()  # Ïó≠ÏàúÏúºÎ°ú Í∑∏ÎûòÌîÑ ÌëúÏãú
plt.show()
````
