 (4) Outlier 확인 (※ 정상적인 범주를 벗어난 Data)

````python
#각 컬럼별 요약 통계량 확인
df.describe()
````

````python
# 컬럼 범위를 선택하여 이상치 여부 확인
selected_columns = df.columns[1:591]  # 0~589 센서데이터 선택

# IQR을 계산하는 함수 정의
def detect_outliers_iqr(series, threshold=1.5):
    """
    IQR (Interquartile Range)을 이용하여 이상치를 탐지하는 함수
    :param series: 데이터 시리즈
    :param threshold: 이상치로 판단하는 기준값 (기본값: 1.5)
    :return: 이상치의 인덱스 리스트
    """
    # 1사분위수와 3사분위수 계산
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    # IQR 계산
    IQR = Q3 - Q1
    # 이상치 경계 계산
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    # 이상치 탐지
    outliers = (series < lower_bound) | (series > upper_bound)
    return outliers

# 선택된 컬럼 범위의 이상치 여부 확인
outliers = df[selected_columns].apply(detect_outliers_iqr)

# 각 컬럼의 이상치 비율 계산
outlier_ratios = outliers.mean() * 100  # 이상치 비율에 100을 곱하여 퍼센트로 변환

# 이상치 비율 출력
print("각 센서데이터 컬럼의 이상치 비율 (%):")
print(outlier_ratios)
````

````python
# 이상치 비율을 내림차순으로 정렬하여 출력
outlier_ratios_sorted = outlier_ratios.sort_values(ascending=False)

# 이상치 비율 출력 (내림차순)
print("각 센서데이터 컬럼의 이상치 비율 (%), 내림차순:")
print(outlier_ratios_sorted)
````

* 모델링을 수행하기 위해 Feature와 예측하고자하는 값인 Y로 데이터를 나눔
* 학습과 예측을 위한 Train / Test set 분할
* Categorical value 인코딩

  - boosting 앙상블 (lightGBM)

```
* Model Selection 단계에서 선정한 모델들을 학습하고 성능을 기록
* 동일한 Data set, 동일한 환경에서 동일한 비교 지표로 성능을 비교
```

````python
from imblearn.over_sampling import SMOTE

# Pass/Fail의 비율 확인
class_distribution = df2['Pass/Fail'].value_counts()
print("Original Class Distribution:")
print(class_distribution)

# 입력 변수와 타겟 변수 분리
X = df2.drop(['Pass/Fail'], axis=1)       # 불량여부와 상관성이 높았던 14개의 센서 데이터 대상으로 모델링 진행
y = df2['Pass/Fail']

# 학습 데이터 및 테스트 데이터로 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2024)

# SMOTE를 사용하여 클래스 불균형 보정
smote = SMOTE(sampling_strategy='auto', random_state=2024)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# 보정된 클래스 분포 확인
resampled_class_distribution = pd.Series(y_train_resampled).value_counts()
print("\nResampled Class Distribution:")
print(resampled_class_distribution)
````

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 로지스틱 회귀 모델 생성
model_lr = LogisticRegression(random_state=10)

# 로지스틱 회귀 모델 학습
model_lr.fit(X_train_resampled, y_train_resampled)

# 모델 평가
y_pred_lr = model_lr.predict(X_test)

# 정확도 및 분류 보고서 출력
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print(f"Accuracy of Logistic Regression Model: {accuracy_lr}")

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_lr))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_lr))
````

````python
from sklearn.ensemble import RandomForestClassifier

# 랜덤 포레스트 모델 생성
model_rf = RandomForestClassifier(n_estimators=100, random_state=10)

# 랜덤 포레스트 모델 학습
model_rf.fit(X_train_resampled, y_train_resampled)

# 모델 평가
y_pred_rf = model_rf.predict(X_test)

# 정확도 및 분류 보고서 출력
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Accuracy of Random Forest Model: {accuracy_rf}")

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf))
````

````python
from lightgbm import LGBMClassifier

# LightGBM 모델 생성
model_lgbm = LGBMClassifier(random_state=10)

# LightGBM 모델 학습
model_lgbm.fit(X_train_resampled, y_train_resampled)

# 모델 평가
y_pred_lgbm = model_lgbm.predict(X_test)

# 정확도 및 분류 보고서 출력
accuracy_lgbm = accuracy_score(y_test, y_pred_lgbm)
print(f"Accuracy of LightGBM Model: {accuracy_lgbm}")

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_lgbm))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_lgbm))
````

 - 모델에서 중요한 변수 확인 (Feature Importance, Permutataion Importance 개념)
 - 사전에 정의한 문제를 해결한 결론 작성 필요

````python
# False Alarm이 가장 낮고 정확도가 높은 LGBM모델로 선택하여 중요 데이터 도출
# 특성 중요도 추출
feature_importances = model_lgbm.feature_importances_

# 특성 중요도를 DataFrame으로 변환하여 출력
importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)
print(importance_df)

# 상위 10개 특성 중요도 추출
top_features = importance_df.head(10)

# 그래프 생성
plt.figure(figsize=(10, 6))
plt.barh(top_features['Feature'], top_features['Importance'], color='lightgrey')
plt.xlabel('Importance')
plt.title('Top 10 Feature Importances')
plt.gca().invert_yaxis()  # 역순으로 그래프 표시
plt.show()
````

````python
from sklearn.inspection import permutation_importance

# Permutation Importance 계산
# 특성값을 순열로 만든 후 모델의 예측 오차 증가량을 계산하여 특성값의 중요도를 측정
result = permutation_importance(model_lgbm, X_test, y_test, n_repeats=10, random_state=42, n_jobs=-1)

# 특성 중요도를 DataFrame으로 변환하여 출력
importance_df_permutation = pd.DataFrame({'Feature': X.columns, 'Importance': result.importances_mean})
importance_df_permutation = importance_df_permutation.sort_values(by='Importance', ascending=False)
print(importance_df_permutation)

# 상위 10개 특성 중요도 추출
top_features_permutation = importance_df_permutation.head(10)

# 그래프 생성
plt.figure(figsize=(10, 6))
plt.barh(top_features_permutation['Feature'], top_features_permutation['Importance'], color='lightgrey')
plt.xlabel('Importance')
plt.title('Top 10 Permutation Feature Importances')
plt.gca().invert_yaxis()  # 역순으로 그래프 표시
plt.show()
````

````python
# Feature Importance와 Permutation Importance를 결합하여 상위 10개의 특성을 추출
combined_importance = importance_df.merge(importance_df_permutation, on='Feature')
combined_importance['Combined_Importance'] = combined_importance['Importance_x'] + combined_importance['Importance_y']
combined_importance = combined_importance.sort_values(by='Combined_Importance', ascending=False)
top_10_features_combined = combined_importance.head(10)

print("Top 10 Features (Combined Importance):")
print(top_10_features_combined)

# 그래프 생성
plt.figure(figsize=(10, 6))
plt.barh(top_10_features_combined['Feature'], top_10_features_combined['Combined_Importance'], color='cornflowerblue')
plt.xlabel('Combined Importance')
plt.title('Top 10 Features (Combined Importance)')
plt.gca().invert_yaxis()  # 역순으로 그래프 표시
plt.show()
````
